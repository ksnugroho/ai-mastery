### Global & Local Scope

```python
# Global Scope
nilai = 100

def cetak():
  print(nilai + 50)

cetak()
```

```python
def cetak():
  # local scope
  nilai_2 = 50

cetak()
print(nilai_2)
```

```python
# mengubah menjadi global scpe

def cetak():
  global nilai_2
  nilai_2 = 50

cetak()
print(nilai_2)
```

# B. Iterators

## 01 Generators

```python
def simple_generators():
  yield 1
  yield 2
  yield 3

# x is a generator object
x = simple_generators()

print('Type of x - ', type(x))
```

```python
# iterating over the generator object
print(next(x))
```

```python
print(next(x))
```

```python
print(next(x))
```

```python
print(next(x))
```

```python
def colors():
  yield "red"
  yield "blue"
  yield "green"
   
next_color =colors()
    
print(next(next_color))
print(next(next_color))
print(next(next_color))
```

## 02 map()

```python
def hitung_kuadrat(a):
	return a*a

daftar_bil = [0, 2, 3, 4, 5]

hasil = map(hitung_kuadrat, daftar_bil)

print(hasil)
print(type(hasil))
print(list(hasil))
```

```python
# using lambda functions with map

daftar_bil = [0, 2, 3, 4, 5]

hasil = map(lambda a:a*a, daftar_bil)

print(list(hasil))
```

## 03 filter()

```python
alfabet = ['a', 'b', 'c', 'd', 'e', 'u', 'k', 'i', 'x', 'y', 'z']

def filter_huruf_vokal(alfabet):
	vokal = ['a', 'i', 'u', 'e', 'o']
	if alfabet in vokal:
		return True
	else:
		return False

vokal_terfilter = filter(filter_huruf_vokal, alfabet)

for i in vokal_terfilter:
	print(i)
```

```python
def bil_genap(a):
  if a%2 == 0:
    return True
  else:
    return False

daftar_bil = [2,3,4,5,6,7,8,9,10]

hasil = filter(bil_genap, daftar_bil)

print(hasil)
print(type(hasil))
print(list(hasil))
```

# C. String Manipulation

```python
string = "Hello World"

print(string.upper())
print(string.lower())
print(string.title())
print(string.capitalize())
print(string.swapcase())
print(string.split(' '))
print(string.find('World'))
print(string.replace('World','Cahyo'))
print(' '.join(reversed(string)))
```

```python
# Python string format method

# default
print('default format: {}, {}, {}'.format('Cahyo','Widya','Roy'))

# positional argument
print('positional order: {1}, {0}, {2}'.format('Cahyo','Widya','Roy'))

# keyword argument
print('keyword order: {z}, {x}, {y}'.format(x='Cahyo',y='Widya',z='Roy'))
```
